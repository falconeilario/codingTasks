```python
# Import libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
```

```python
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
```

### 1. Boxplot

```python
# Boxplot for revs per mile
```

```python
selected_manufacturers = car_data[car_data["Manufacturer"].isin(["Audi", "Hyundai", "Suzuki", "Toyota"])]
revs_per_mile_boxplot = sns.boxplot(data=selected_manufacturers, x="Rev.per.mile", y="Manufacturer")
revs_per_mile_boxplot.set(xlabel ="Rev. per mile", ylabel = "Manufacturer", title ="Rev. per mile comparison")
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**


Toyota has the car with the highest revs per mile.

### 2. Histogram

```python
# Histogram of MPG
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**

```python
features_to_plot = pd.DataFrame({'MPG.highway': car_data['MPG.highway'],
                                 'MPG.city': car_data['MPG.city']})
sns.histplot(data=features_to_plot, x="MPG.city", element="step")
sns.histplot(data=features_to_plot, x="MPG.highway", element="step")
```

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
```

```python
sns.lineplot(data=car_data, x="Wheelbase", y="Turn.circle")
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  


When the wheelbase gets larger the turning circle increases.

### 4. Barplot

```python
# Barplot for mean horsepower
```

```python
sns.barplot(data=car_data, x="Horsepower", y="Type")
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**

No, a larger car does not mean more horsepower
